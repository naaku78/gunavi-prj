```python
# Rearrange in 
tickers = ['AAPL','AMZN','BA','F','GE','MMM', 'MSFT','PFE','TSLA','TWTR','XOM']
```

```python
stocks_data = pd.concat([AAPL,AMZN,BA,F,GE,MMM, MSFT,PFE,TSLA,TWTR,XOM],axis=1, keys=tickers)
stocks_data
```

```python
#set the column names levels
stocks_data.columns.names = ['Stock Ticker', 'Stock Info']
```

```python
#create a return dataframe to contain all stocks returns
returns = pd.DataFrame()
```

```python
#Create a loop to return stock ticker columns in dataframe
for tick in tickers:
    returns[tick+' Return'] = stocks_data[tick]['Close'].pct_change()
    
```
